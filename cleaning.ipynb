```python
import pandas as pd
import numpy as np

# 1. LOAD DATA - Ensure this filename matches your uploaded file exactly
filename = 'Student_Budget_Management_Predictor.xlsx'
df = pd.read_excel(filename)

# 2. DEFINE CLEANING FUNCTION
def map_range(val):
    if pd.isna(val) or 'No' in str(val) or '₹0' in str(val) or 'rent' in str(val).lower(): 
        return 0
    
    # Remove currency symbols and commas
    val = str(val).replace('₹', '').replace(',', '').strip()
    
    # Extract numbers from various formats (Below, Above, or Ranges)
    if 'Below' in val:
        num = float(''.join(filter(str.isdigit, val)))
        return num * 0.5  # Use 50% of the 'Below' value as a midpoint
    elif 'Above' in val:
        num = float(''.join(filter(str.isdigit, val)))
        return num * 1.2  # Use 120% of the 'Above' value as a midpoint
    elif '–' in val or '-' in val:
        sep = '–' if '–' in val else '-'
        parts = val.split(sep)
        low = float(''.join(filter(str.isdigit, parts[0])))
        high = float(''.join(filter(str.isdigit, parts[1])))
        return (low + high) / 2
    return 0

# 3. SELECT COLUMNS
money_cols = [
    '2. What is your average monthly income or allowance? ', 
    '4. What is your monthly rent / hostel fee? ',
    '5. How much do you spend on food per month? ',
    '6. What are your monthly transportation expenses? ',
    '7. How much do you spend on mobile recharge and internet per month? ',
    '8. What are your average education-related expenses per month? \n(Books, printing, subscriptions, etc.)',
    '9. How much do you spend monthly on eating out or food delivery? ',
    '10. How much do you spend monthly on shopping and entertainment?\n (Clothes, gadgets, movies, games, etc.) ',
    '13. How much do you save on average per month? '
]

# 4. EXECUTE CLEANING
# This is where your error was—df is now defined above in step 1
for col in money_cols:
    df[col] = df[col].apply(map_range)

# Clean the Target column
df['target'] = df['16. Do you usually run out of money before the end of the month? '].map({'Yes': 1, 'No': 0})

# 5. SAVE CLEANED DATA
df.to_csv('cleaned_student_data.csv', index=False)

print(df[['2. What is your average monthly income or allowance? ', 'target']].head())
```
